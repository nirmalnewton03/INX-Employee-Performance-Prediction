# **Project Summary**  
## **INX Future Inc Employee Performance - Project**  

The **Data Science Project** presented here focuses on analyzing employee performance. The primary goal is to predict employee performance ratings based on various features such as:  
- `TotalWorkExperienceInYears`  
- `EmpDepartment`  
- `Gender`  
- `ExperienceYearsInCurrentRole`, among others.

---

### **Goals and Insights**
1. **Department-Wise Performance Analysis**  
   Understand how performance varies across different departments.  
   
2. **Top 3 Important Factors Influencing Employee Performance**  
   Identify key factors that significantly impact employee performance.  
   
3. **Performance Prediction Model**  
   Develop a machine learning model to predict employee performance based on input factors. This will aid in hiring decisions.  

4. **Actionable Recommendations**  
   Provide recommendations to improve employee performance based on analytical insights.

---

### **Key Methodologies**
1. **Correlation Analysis**  
   - Evaluated relationships between features and performance rating using correlation matrices to identify strong associations.

2. **Machine Learning Model**  
   - **Best Model:** `RandomForestClassifier`  
   - **Accuracy:** **92.08.00%**  
   - **Justification:** High performance, robustness to overfitting, and ability to handle both categorical and numerical data.  
   - **Feature Importance:** Random Forest also provided clear insights into which features most influence performance.

3. **Data Preprocessing**  
   - **Label Encoding:** Applied to ordinal categorical features.  
   - **One-Hot Encoding (OHE):** Used for nominal categorical variables to make them suitable for machine learning models.  
   - **Outlier Handling:** Detected and treated using statistical techniques to ensure data quality.
   - **SMOTE (Synthetic Minority Over-sampling Technique)** was implemented to address class imbalance in the target variable.
   - **PCA (Principal Component Analysis)** was explored for dimensionality reduction; however, it resulted in **lower accuracy**. Therefore, PCA was not used in the final model as performance was **better without it**.


4. **Visualization Techniques**  
   - Used **distribution plots**, **bar charts**, and **correlation heatmaps** to uncover trends, relationships, and data patterns visually.
