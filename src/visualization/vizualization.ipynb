## 3.DATA ANALYSIS USING VISUALIZATION 

### UNIVARIATE ANALYSIS FOR NUMERICAL FEATURES

```python
sns.set_style("darkgrid")
num_cols = ['Age', 'DistanceFromHome', 'EmpHourlyRate', 'NumCompaniesWorked', 
            'EmpLastSalaryHikePercent', 'TotalWorkExperienceInYears', 'TrainingTimesLastYear', 
            'ExperienceYearsAtThisCompany', 'ExperienceYearsInCurrentRole', 
            'YearsSinceLastPromotion', 'YearsWithCurrManager']

plt.figure(figsize=(30, 30), facecolor='White')
plot_number = 1

for s in num_cols:
    if plot_number <= 11:
        sp = plt.subplot(4, 3, plot_number)
        sns.histplot(x=s, data=data, kde=True, edgecolor="black", alpha=0.8) 
        
        plt.xlabel(s.title(), fontsize=20, color='black')
        plt.ylabel('Count', fontsize=20, color='black')
        plt.title(f"Distribution of {s}", fontsize=22, fontweight='bold', color='black')

        plt.grid(axis='y', linestyle='--', alpha=0.5)

    plot_number += 1

plt.tight_layout()
plt.show()
```

### UNIVARIATE ANALYSIS FOR CATEGORICAL FEATURES

```python
sns.set_style("darkgrid")
cat_cols = ['Gender', 'EducationBackground', 'MaritalStatus', 'EmpDepartment', 'EmpJobRole', 'BusinessTravelFrequency','EmpEducationLevel', 'EmpEnvironmentSatisfaction', 'EmpJobInvolvement', 'EmpJobLevel',
       'EmpJobSatisfaction','OverTime','EmpRelationshipSatisfaction','EmpWorkLifeBalance','Attrition']

plt.figure(figsize=(30, 30), facecolor='White')
plot_number = 1

for s in cat_cols:
    if plot_number <= 11:
        sp = plt.subplot(5, 3, plot_number)
        sns.countplot(x=s, data=data,palette='viridis' ,edgecolor="black", alpha=0.8) 
        
        plt.xlabel(s.title(), fontsize=20, color='black')
        plt.ylabel('Count', fontsize=20, color='black')
        plt.title(f"Distribution of {s}", fontsize=22, fontweight='bold', color='black')
        plt.xticks(rotation=45)

        plt.grid(axis='y', linestyle='--', alpha=0.5)

    plot_number += 1

plt.tight_layout()
plt.show()
```

```python
data['OverTime'] = data['OverTime'].map({'Yes': True, 'No': False})
data['Attrition'] = data['Attrition'].map({'Yes': True, 'No': False})

plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='OverTime', hue='Attrition')

plt.xlabel("Overtime", fontsize=13)
plt.ylabel("Number of Employees", fontsize=13)
plt.title("Attrition by Overtime", fontsize=16, fontweight='bold')

plt.xticks([0, 1], ['No Overtime', 'Overtime'], fontsize=12)
plt.legend(title='Attrition', labels=['Did Not Leave', 'Left'], fontsize=12)

sns.despine(left=True, bottom=False)
plt.tight_layout()
plt.show()

overtime_attrition = data.groupby('OverTime')['Attrition'].value_counts().unstack(fill_value=0)
overtime_attrition['Total'] = overtime_attrition.sum(axis=1)
overtime_attrition['Attrition Rate (%)'] = (overtime_attrition[True] / overtime_attrition['Total']) * 100

plt.figure(figsize=(6, 4))
sns.barplot(x=overtime_attrition.index, y=overtime_attrition['Attrition Rate (%)'])

plt.xlabel("Overtime", fontsize=13)
plt.ylabel("Attrition Rate (%)", fontsize=13)
plt.title("Attrition Rate by Overtime", fontsize=16, fontweight='bold')
plt.xticks([0, 1], ['No Overtime', 'Overtime'], fontsize=12)

sns.despine(left=True, bottom=False)
plt.tight_layout()
plt.show()
```

##  Key Insights from Visualizations


1. **Training Hours vs Performance**  
   **Training improves performance**, especially for underperformers, but **excessive training yields diminishing returns**.

2. **Department-wise Attrition**  
   **Sales department** experiences the **highest employee attrition** among all departments.

3. **Performance Ratings**  
   Most employees have a **Performance Rating of 3**, with relatively **few high performers (Rating 4 or 5)**.

4. **Hourly Pay vs Performance**  
   **Hourly wages** have **no clear impact** on performance. **Higher pay** doesn't guarantee better performance, nor does **lower pay** imply underperformance.

5. **Work-Life Balance vs Productivity**  
   A **better work-life balance** is generally associated with **higher productivity** and performance.

6. **Department Size**  
   The **Sales** and **R&D** departments have the **largest employee base**, while **HR** and **Legal** have the **fewest employees**.

7. **Employee Demographics**  
   Majority of the workforce falls into the **young to mid-career** age range.

8. **Overtime vs Attrition**  
   **Overtime** has **minimal impact** on attrition rates, suggesting other factors are more influential.

9. **Performance vs Job Satisfaction Factors**  
   **Attrition**, **Job Satisfaction**, and **Work-Life Balance** do **not significantly affect performance ratings** in the current dataset.

10. **Salary Hike vs Performance**  
   Employees receiving a **salary hike of more than 20%** are typically **high performers**, suggesting that **better performance is rewarded** with larger hikes.

11. **Marital Status & Work-Life Balance**  
   **Married employees** tend to report a **better work-life balance** compared to unmarried employees.

12. **Environment Satisfaction**  
   Employees with **higher Environment Satisfaction** tend to **perform better**, indicating a strong relationship between workplace environment and output.
