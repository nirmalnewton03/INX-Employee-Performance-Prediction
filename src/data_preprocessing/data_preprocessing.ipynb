## 3.2. Encoding

### 3.2.1. One-Hot Encoding

```python
ohe_columns = ["Gender", "OverTime", "Attrition"]
data= pd.get_dummies(df, columns=ohe_columns, drop_first=True)

for col in data.columns:
    if any(cat in col for cat in ohe_columns):
        data[col] = data[col].astype(int)
        
data.head()
```

### 3.2.2. Label Encoding

```python
le_columns=['EmpDepartment','MaritalStatus','EmpJobRole','BusinessTravelFrequency','EducationBackground']
label_encoders = {}

for col in le_columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])  
    label_encoders[col] = le

data.head()
```

## 3.3. Outliers Handling

```python
outlier_cols = ['TotalWorkExperienceInYears','ExperienceYearsInCurrentRole', 'YearsWithCurrManager', 
                'YearsSinceLastPromotion', 'ExperienceYearsAtThisCompany', 'TrainingTimesLastYear','NumCompaniesWorked']

def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
        df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
    
    return df

data = remove_outliers(data, outlier_cols)
```

```python
def count_outliers(data, columns):
    outlier_counts = {}
    
    for col in columns:
        Q1 = data[col].quantile(0.25) 
        Q3 = data[col].quantile(0.75)  
        IQR = Q3 - Q1  
        
        lower_bound = Q1 - 1.5 * IQR  
        upper_bound = Q3 + 1.5 * IQR 
        
        outliers = ((data[col] < lower_bound) | (data[col] > upper_bound)).sum()  # Count of outliers
        outlier_counts[col] = outliers

    return pd.DataFrame.from_dict(outlier_counts, orient='index', columns=['Number of Outliers'])

outlier_summary = count_outliers(data, num_cols)
print(outlier_summary)
```

## 3.4. Handling Skewness

```python
num_cols = ['Age', 'DistanceFromHome', 'EmpHourlyRate', 'NumCompaniesWorked', 
            'EmpLastSalaryHikePercent', 'TotalWorkExperienceInYears', 'TrainingTimesLastYear', 
            'ExperienceYearsAtThisCompany', 'ExperienceYearsInCurrentRole', 
            'YearsSinceLastPromotion', 'YearsWithCurrManager']

num_features = len(num_cols)
rows = int(np.ceil(num_features / 3))  

plt.figure(figsize=(18, 6 * rows))  
for i, col in enumerate(num_cols, 1):
    plt.subplot(rows, 3, i) 
    sns.histplot(data[col], bins=30, kde=True)  
    plt.title(f'Distribution of {col}', fontweight='bold', fontsize=15)
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.subplots_adjust(hspace=0.3)
plt.show()
```

```python
#checking for skewness
for i in num_cols:
    skewness_values=data[i].skew()
    print(f'{i}: {skewness_values}')
```
